```python
def forward_pass(X, pi, A, mu, sigma):
    T = len(X)
    N = len(pi)
    alpha = np.zeros((T,N))
    c = np.zeros(T)

    b0 = norm.pdf(X[0], mu, sigma)
    alpha[0] = pi * b0
    c[0] = alpha[0].sum()
    if c[0] <= 0:
        return None, None, -np.inf 
    alpha[0] /= c[0]

    #recursion
    for t in range (1,T):
        bt = norm.pdf(X[t], mu, sigma)
        alpha[t] = (alpha[t-1] @ A) * bt
        c[t] = alpha[t].sum()
        if c[t] <= 0:
            return None, None, -np.inf
    alpha[t] /= c[t]

    logL = np.sum(np.log(c))
    return alpha, c, logL

```
